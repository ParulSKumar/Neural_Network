```


import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/ParulSKumar/Neural_Network/main/LBW_Dataset.csv')
print(data.head())
#no values missing from column community
# For age taking mean of all ages to fill in missing values
# Or could take mean of ages of that specific community

data['Age']=data['Age'].fillna(data['Age'].mean())
data['Age']=data['Age'].astype(int)
print(data.tail())

# Similarly taking mean for filling missing weights

data['Weight']=data['Weight'].fillna(data['Weight'].mean())
data['Weight']=data['Weight'].astype(int)
print(data.tail())

# Columns phase, residence should be int not float, they are essentially categorical

#For missing delivery phase filling 1, seems more probable that a normal phase was not recorded as compared to an abnormal one
data['Delivery phase']=data['Delivery phase'].fillna(1)
data['Delivery phase']=data['Delivery phase'].astype(int)
print(data.tail())

#For missing HB considering the most frequently occuring value
data['HB']=data['HB'].fillna(data['HB'].median())
print(data.tail())

#For missing residence values filling 1, more probable
data['Residence']=data['Residence'].fillna(1)
data['Residence']=data['Residence'].astype(int)
print(data.tail())

#All values in education are 5 it will be fine to drop this column doesn't give us any useful information
del data['Education']
print(data.head())

#For blood pressure considering the most frequently occuring value
data['BP']=data['BP'].fillna(data['BP'].median())
print(data.tail())
```
